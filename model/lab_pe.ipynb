```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('pe_train2.csv')
data
```

```python
# null 값이 있는 컬럼은?
for column, isnull in data.isnull().sum().items():
    if isnull: print(column)
```

```python
# class 정보
data['class'].value_counts().plot(kind='bar')
plt.title('class')
plt.show()
```

```python
drop_list_index = [8, 11, 16, 17, 18, 20, 23, 25, 27, 28, 31, 93, 123]
drop_list = ['FH_char1',
             'FH_char4',
             'FH_char9',
             'FH_char10', 
             'FH_char11',
             'FH_char13', 
             'OH_DLLchar1', 
             'OH_DLLchar3', 
             'OH_DLLchar5', 
             'OH_DLLchar6', 
             'OH_DLLchar9', 
             'VersionInformationSize',
             'pe_driver']

for drop_column in drop_list:
    data = data.drop([drop_column], axis=1)
```

```python
data.to_csv('pe_train2_preprocess1.csv')
```

# 분석

```python
columns = data.drop(['filename', 'md5', 'CheckSum', 'CreationYear', 'class'], axis=1).columns

for i in range(len(columns)):
    data[columns[i]].value_counts().plot(kind='bar')
    plt.title(str(i) + ' ' + columns[i])
    plt.show()
```
