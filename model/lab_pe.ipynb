```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('pe_train.csv')
data.head(30)
```

```python
data.info()
```

```python
# null 값이 있는 컬럼은?
for column, isnull in data.isnull().sum().items():
    if isnull: print(column)
```

```python
columns = data.columns

for i in range(1, len(columns)):
    data[columns[i]].value_counts().plot(kind='bar')
    plt.title(str(i) + ' ' + column)
    plt.show()
```

```python
drop_list = [8, 11, 16, 17, 18, 20, 23, 25, 27, 28, 31, 57, 58, 93, 98]

columns = data.columns
for drop_index in drop_list:
    data = data.drop([columns[drop_index]], axis=1)
```
